```python
# Задание task_03_02_01

# Выполнила: Александрова Екатерина.
# Группа: ЦИБ - 251

def foo(a):  # создаём функцию foo, которая принимает список a
    """
    Сортировка пузырьком.

    Худшая временная сложность: O(n²). (выполняется два вложенных цикла)
    Лучшая временная сложность: O(n²).
    Сложность по памяти: O(1). (дополнительные структуры данных не используются)
    """
    for i in range(len(a), 0, -1):  # внешний цикл: количество проходов по списку
        for j in range(1, i):  # внутренний цикл: сравниваем соседние элементы
            if a[j - 1] > a[j]:  # если левый элемент больше правого
                a[j - 1], a[j] = a[j], a[j - 1]  # меняем элементы местами

    return a  # возвращаем отсортированный список


a = [1, 2, 3, 4, 2, 1, 3, 4, 3, 5, -2, -2, 2, -2, 65, 2, 2]
# создаём список чисел

foo(a)  # вызываем функцию сортировки

print(a)  # выводим отсортированный список на экран
```
